```python
lst = [int(s) for s in input().split()]
valore = int(input())
lst2 = []

for elemento in lst:
  n2 = elemento * valore
  lst2.append(n2)

print(lst)
print(lst2)
```

```python
listanum = [int(ele) for ele in input ("Inserire i numeri: ").split()]
valore = int(input("Inserire un valore: "))
listafinale = []

for numero in listanum:
  if numero%valore == 0:
    listafinale.append(numero)

print(listanum)
print(listafinale)
```

```python
listanum = [int(ele) for ele in input ("Inserire i numeri: ").split()]
valore = int(input("Inserire un valore: "))
listafinale = [i for i in listanum if i%valore==0]
print(listanum)
print(listafinale)
```

```python
import math
def superficie (r):
  return r**2*math.pi
superficie(int(input()))
```

```python
def volume (r,h):
  return superficie(r)*h
print(volume(5,10))
```

```python
numeri = [int(s) for s in input().split()]
numeriDec = list(map(lambda x:x-1, numeri))
print(numeri)
print(numeriDec)
```

```python
numeri = [int(s) for s in input().split()]
numeriPari = list(filter(lambda x:x%2==0, numeri))
print(numeri)
print(numeriPari)
```

```python
def potenza (base, esponente):
  if esponente == 0:
    return 1
  else:
    return base*potenza(base,esponenete-1)
  
print(potenza(5,4))
```
