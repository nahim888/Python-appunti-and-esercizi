1. scrivere un programma che data una data nella forma gg/mm/aaaa visualizzi separatamente il giorno, il mese e l'anno

```python
data = input("Inserire una data: ")
if len(data)==10 and data[2]=="/" and data[5]=="/":
  print("Il giorno è: " + data[0:2])
  print("Il mese è: " + data[3:5])
  print("L'anno è: " + data[6:10])
```

2. scrivere un programma che permetta di stabilire se un numero dispari è divisibile per 5. L'utente inserisce un numero e il programma, una volta controllato che sia dispari, controlla anche se è divisibile per 5.

```python
n = int(input("Inserire un numero: "))
if n%2==0:
  print("Il numero non è dispari ")
else:
  if n%5==0:
    print("Il numero è dispari ed è divisibile per 5")
  else:
    print("Il numero è dispari ma non è divisibile per 5")
```

3. scrivere un programma per verificare se una parola comincia e finisce con la stessa lettera

```python
parola = input("Inserire una parola: ")
if parola[0] == parola[-1:]:
  print("La parola comincia e finisce con la stessa lettera")
else:
  print("La parola non comincia e finisce con la stessa lettera")
```

4. scrivere un programma che permetta di caricare una lista di parole e che elimini dalla lista tutte le parole che iniziano con una lettera inserita dall'utente

```python
lst = [str(elemento) for elemento in input("Inserisci le parole: ").split()]
lettera = input("Inserire una lettera: ")
lst2 = lst[:]
for elemento in lst:
  if elemento[0]==lettera:
    lst2.remove(elemento)
print(lst)
print(lst2)
```

5. scrivere una funzione che data una parola restituisca la stessa parola senza le vocali




```python
def rimuoviVocali(string):
    vocali = ['a','e','i','o','u']
    r = [i for i in string if i.lower() not in vocali]
    r = ''.join(r)
    print(r)

parola = input("Inserire una parola: ")
rimuoviVocali(parola)
```

6. scrivere un programma che utilizzando la funzione precedente elimini da una frase composta da più parole tutte le vocali

```python
frase = input("Inserire una frase: ")
rimuoviVocali(frase)
```

7. scrivere un programma che utilizzando una lambda function  sostituisca ad ogni elemento di una lista di parole la lunghezza delle parole stesse. Ad esempio se la lista fosse [‘mamma’, ‘stella’, ‘casa’] la lista finale dovrebbe essere [5, 5, 4]


```python
lst =  [str(elemento) for elemento in input("Inserisci le parole: ").split()]
lstf = list(map(lambda x: len(x), lst))
print(lst)
print(lstf)
```

8. scrivere un programma che utilizzando una lambda function elimini da una lista di parole tutte quelle più corte di un valore inserito dall’utente.  Ad esempio se la lista fosse [‘tre’, ‘stella’, ‘casa’] e l’utente inserisse il valore 5, la lista finale dovrebbe essere [‘stella’]


```python
lst =  [str(elemento) for elemento in input("Inserisci le parole: ").split()]
n = int(input("Inserire un numero: "))
lstf = list(filter(lambda x: len(x)>n, lst))
print(lst)
print(lstf)
```

9. MilanoSport propone corsi di nuoto e vuole memorizzare i nomi dei corsi e i cognomi delle persone che partecipano a quel corso. Caricare in una struttura dati apposita i nomi dei corsi (solo due nomi) e i cognomi dei partecipanti (solo due cognomi) e permettere di avere i cognomi dei partecipanti di un determinato corso inserito dall’utente. 


```python
MilanoSport = [{"nome corso" : "corso 1", "Primo partecipante" : "A", "Secondo partecipante" : "C"},
                 {"nome corso" : "corso 2", "Primo partecipante" : "B", "Secondo partecipante" : "D"}]

corso = input("Inserisci il corso: ")

for r in MilanoSport:
  if r["nome corso"] == corso:
    print("Il primo partecipante è: " + r["Primo partecipante"] + " " + "Il secondo partecipante è: " + r["Secondo partecipante"])
```
