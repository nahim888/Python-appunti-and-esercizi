```python
import pandas as pd
import matplotlib.pyplot as plt
```

crei un dataframe contenente le informazioni riportate nella tabella sottostante, relativa al numero di ordini (order_numbers) seguiti da vari commessi (last_name) di un negozio.


```python
d = {"last_name" : ["Boyer", "Copeland", "Daniel", "Serrano", "Terrell", "Vargas"],
     "order_numbers" : [185,50,196,58,24,28]
     }
df = pd.DataFrame(d)
df
```

visualizzi un grafico a barre verticali per rappresentare i dati. Il risultato deve essere questo:


```python
data = df.groupby('last_name').sum()
names = list(data.index)
values = list(data['order_numbers'])
fig = plt.figure()
ax = plt.axes()
ax.bar(names, values)
ax.set_xlabel('x') #per mettere un nome all'asse delle ascisse
ax.set_ylabel('y') #per mettere un nome all'asse delle ordinate
fig.suptitle('Istogramma') #per mettere un titolo alla tabella
plt.show()
```

visualizzi lo stesso grafico ma con altri colori delle barre, un titolo appropriato al grafico e i nomi degli assi.

```python
data = df.groupby('last_name').sum()
names = list(data.index)
values = list(data['order_numbers'])
fig = plt.figure()
ax = plt.axes()
ax.bar(names, values, color = "c")
ax.set_xlabel('x')
ax.set_ylabel('y')
fig.suptitle('Istogramma dipendenti')
plt.show()
```

Visualizzare lo stesso grafico ma con le barre in ordine decrescente


```python
data = df.groupby('last_name').sum().sort_values("order_numbers", ascending=False)
names = list(data.index)
values = list(data['order_numbers'])
fig = plt.figure()
ax = plt.axes()
ax.bar(names, values, color = "c")
ax.set_xlabel('x')
ax.set_ylabel('y')
fig.suptitle('Istogramma dipendenti')
plt.show()
```

visualizzi un grafico a barre orizzontali per rappresentare i dat. Il risultato deve essere uguale a questo:

crei un nuovo dataframe contenente la percentuale di ordini consegnati seguiti da ogni componente dello staff. Visualizzare il cognome del componente dello staff e la percentuale di ordini. Ordinare in ordine alfabetico sul cognome del componente dello staff. Il risultato deve essere questo:


```python
df2 = pd.DataFrame(df["last_name"])
df2["orders_perc"] = df["order_numbers"]/df["order_numbers"].sum()*100
df2.sort_values("last_name", ascending=True)
```

visualizzi un grafico a torta per rappresentare i dati estratti. Il risultato deve essere questo:

```python
cols = ['c','m','r','g','b','y']

fig = plt.figure(figsize=[7,7])
ax = plt.axes()
ax.pie(df2["orders_perc"], labels=df2["last_name"], colors=cols, startangle=90, autopct='%1.1f%%')

plt.show()
```

visualizzi lo stesso grafico ma con altri colori delle percentuali e un titolo appropriato al grafico


```python
cols = ['c','m','r','g','b','y']

fig = plt.figure(figsize=[7,7])
ax = plt.axes()
ax.pie(df2["orders_perc"], labels=df2["last_name"], colors=cols, startangle=90, autopct='%1.1f%%')
fig.suptitle('Grafico Percentuali Dipendenti', color='k')

plt.show()
```
