```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Creare un DataFrame da questo file json ed assegnarlo ad una variabile df.


```python
df = pd.read_json("https://raw.githubusercontent.com/CappuCitti/Python-appunti-ed-esercizi/main/MatPlotLib/Data/sample_chat_data.json")
df
```

```python
df.info()
```

Creare un grafico a barre che mostri il numero di messaggi inviato da ogni utente in ordine

```python
msg_num = df.groupby("user_id", as_index=False).count().sort_values("message_length", ascending = False)
msg_num
```

```python
#msg_num = df.groupby("user_id")["message_length"].count().sort_values(ascending = False)
#msg_num
```

```python
msg_num["message_length"].mean()
```

```python
msg_num[msg_num["message_length"] > msg_num["message_length"].mean()]["user_id"]
#msg_num.loc[msg_num["message_length"] > msg_num["message_length"].mean(), "user_id"]
```

```python
msg_num[msg_num["message_length"] == msg_num["message_length"].max()]["user_id"]
#msg_num.loc[msg_num["message_length"] == msg_num["message_length"].max(), "user_id"]
```

```python
plt.rcParams.update({"font.size" : 20})
fig = plt.figure(figsize=(9.6,7.2))
ax = plt.axes()
fig.autofmt_xdate(rotation=45)
ax.bar(msg_num["user_id"], msg_num["message_length"], color = "lightgreen")
fig.suptitle("Messaggi")
ax.set_xlabel("ID Utenti")
ax.set_ylabel("Numero messaggi")
ax.grid()
plt.show()
```

```python
df["full_date"] = pd.to_datetime(df["full_date"])
df["full_date"].dt.date
```

```python
plt.rcParams.update({"font.size" : 12})
data = df.groupby(df["full_date"].dt.date).count()
fig = plt.figure(figsize=(10,10))
ax = plt.axes()
fig.suptitle("Grafico a torta")
ax.pie(data["user_id"], labels = data.index, colors=["red","orange","yellow","green","blue","indigo","purple"], startangle=90, explode = (0,0,0,0.5,0,0,0), autopct='%1.1f%%')
plt.show()
```

```python
data = df.groupby("user_id").agg({"message_length" : "count", "has_media" : "sum", "has_links" : "sum"}).sort_values("message_length", ascending = False)
labels = data.index
msgs = np.subtract(data["message_length"], np.add(data["has_media"], data["has_links"]))
msgs_links = data["has_links"]
msgs_media = data["has_media"]

fig = plt.figure(figsize=(9.6,7.2))
ax = plt.axes()
fig.autofmt_xdate(rotation=45)
#fig.autofmt_xdate(rotation="vertical")

ax.bar(labels,msgs) #color = "#f00"
ax.bar(labels,msgs_links, bottom=msgs) #color = "#00f"
ax.bar(labels,msgs_media, bottom=np.add(msgs, msgs_links)) #color = "#0f0"
#ax.bar(labels,msgs_media, bottom=(msgs+msgs_links))

plt.show()
```
