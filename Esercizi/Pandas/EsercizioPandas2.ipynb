```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/anagrafica-vaccini-summary-latest.csv"

df = pd.read_csv(url)
df
```

Visualizzare il numero di vaccinati nella fascia 12-19

```python
df.loc[df["fascia_anagrafica"] == "12-19", "totale"]
```

Visualizzare il numero di vaccinati maschi e femmine

```python
df.loc[df["fascia_anagrafica"] == "12-19", ["sesso_maschile", "sesso_femminile"]]
```

Visualizzare le fasce di età con più di 10mln di vaccinazioni

```python
df.loc[df["totale"] >= 10000000, "fascia_anagrafica"]
```

Visualizzare le fasce di età in cui sono state vaccinate fra i 10 e i 20 milioni di persone

```python
df.loc[(df["totale"] > 10000000) & (df["totale"] < 20000000), "fascia_anagrafica"]
```

Aggiungere una colonna "percentuale_sesso_maschile" dove inserire la percentuale di maschi vaccinati sul totale

```python
df["percentuale_sesso_maschile"] = (df["sesso_maschile"]/df["totale"])*100
df
```

Fare la stessa cosa per le femmine

```python
df["percentuale_sesso_femminile"] = (df["sesso_femminile"]/df["totale"])*100
df
```

Visualizzare le fasce di età in cui la percentuale di maschi vaccinati è maggiore della percentuale di femmine vaccinate

```python
df.loc[(df["sesso_maschile"] > df["sesso_femminile"]),"fascia_anagrafica"]
```

Visualizzare le fasce di età dove la dose aggiuntiva è minore della dose booster

```python
df.loc[(df["dose_aggiuntiva"] < df["dose_booster"]),"fascia_anagrafica"]
```
