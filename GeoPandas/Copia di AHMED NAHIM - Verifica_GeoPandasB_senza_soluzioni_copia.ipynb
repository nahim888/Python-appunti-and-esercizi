```
# caricare il geodataframe dei quartieri ds964_nil_wm.zip
```

```
quartieri = geopandas.read_file("/content/drive/MyDrive/Colab Notebooks/geopandas-tutorial-master/ds964_nil_wm.zip")
quartieri
```

```
# visualizza la mappa dei quartieri e delle piste ciclabili
```

```
ax = bike_ciclabili.to_crs(epsg=3857).plot(figsize = (20,20))
quartieri.to_crs(epsg=3857).plot(ax=ax, edgecolor = "k", facecolor = "none")
ctx.add_basemap(ax)
```

```
# creare una geoserie con le informazioni sulla ciclabile di via Melchiorre Gioia e visualizzarla
```

```
a = bike_ciclabili[bike_ciclabili["anagrafica"] == "VIA MELCHIORRE GIOIA"]
a.plot()
```

```
# trovare tutti i quartieri attraversati da questa pista ciclabile
```

```
b = quartieri[quartieri.intersects(a.unary_union)]
b
```

```
# visualizzare su una mappa la pista ciclabile e i quartieri attraversati (utilizzare i colori presenti nell'esempio)
```

```
ax = b.to_crs(epsg=3857).plot(edgecolor = "k", facecolor = "yellow", figsize = (20,20))
a.to_crs(epsg=3857).plot(ax=ax)
ctx.add_basemap(ax)
```

```
# caricare il geodataframe delle fontanelle Fontanelle.zip
```

```
fontanelle = geopandas.read_file("/content/drive/MyDrive/Colab Notebooks/geopandas-tutorial-master/Fontanelle.zip")
fontanelle
```

```
# visualizzare il geodataframe su una mappa
```

```
ax = fontanelle.to_crs(epsg=3857).plot(figsize = (20,20))
ctx.add_basemap(ax)
```

```
# visualizzare su una mappa le fontanelle e i quartieri in cui si trovano.
# non visualizzare i quartieri in cui non ci sono fontanelle
```

```
ax = quartieri[quartieri.intersects(fontanelle.unary_union)].to_crs(epsg=3857).plot(figsize = (20,20), edgecolor="k", facecolor="none", linewidth = 3)
fontanelle.to_crs(epsg=3857).plot(ax=ax)
ctx.add_basemap(ax)
```

```
# colorare i quartieri in modo proporzionale alla loro dimensione
```

```
ax = quartieri.to_crs(epsg=3857).plot(column="Shape_Area", figsize = (20,12), legend=True)
```

```
# colorare i quartieri in modo proporzionale al numero di fontanelle presenti (giallo molte fontanelle, blu scuro poche fontanelle)
# non evidenziare i quartieri senza fontanelle
```

```
# visalizzare la mappa precedente facendo in modo che i quartieri con fontanelle siano colorati con gradazioni di rosso
# e quelli senza fontanelle con gradazioni di grigio
```

```
# trovare il numero di fontanelle presenti nel quartiere Niguarda
```

```
niguarda = quartieri[quartieri["NIL"].str.contains("NIGUARDA")]
fontanelleniguarda = fontanelle[fontanelle.within(niguarda.geometry.squeeze())]
fontanelleniguarda
```

```
# visualizzare su una mappa le fontanelle presenti nel quartiere Niguarda
```

```
ax = niguarda.to_crs(epsg=3857).plot(figsize = (10,10), facecolor = "None")
fontanelleniguarda.to_crs(epsg=3857).plot(ax=ax)
ctx.add_basemap(ax)
```
