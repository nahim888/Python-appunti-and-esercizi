1. visualizzi i parchi di Milano e i comuni che vi fanno parte (vedi immagine allegata)


```python
milano = comuni[comuni.COMUNE == "Milano"]
parchimilano = parchi[parchi.intersects(milano.geometry.squeeze())]
comuniparchi = comuni[comuni.intersects(parchimilano.unary_union)]

ax = comuniparchi.to_crs(epsg=3857).plot(facecolor = "None", edgecolor = "k", figsize = (20,12))
parchimilano.to_crs(epsg=3857).plot(ax=ax, facecolor = "green", alpha = 0.5)
ctx.add_basemap(ax)
```

Visualizzare i comuni confinanti con Milano

```python
comuniconfinanti = comuni[comuni.touches(milano.geometry.squeeze())]
comuniconfinanti.plot()
```

Visualizzare i comuni confinanti ai comuni confinanti con Milano

```python
comuniconfinatiaicomuniconfinanti = comuni[comuni.touches(comuniconfinanti.unary_union)]
comuniconfinatiaicomuniconfinanti.plot()
```

Visualizzare il comune confinante a Milano più esteso

```python
comunemax = comuniconfinanti[comuniconfinanti["Shape_Area"] == comuniconfinanti["Shape_Area"].max()]
comunemax
```

Visualizzare il comune confinante a Milano meno esteso

```python
comunemin = comuniconfinanti[comuniconfinanti["Shape_Area"] == comuniconfinanti["Shape_Area"].min()]
comunemin
```

Visualizzare su una mappa il comune di Milano, il comune confinante più esteso e quello meno esteso

```python
ax = milano.to_crs(epsg=3857).plot(facecolor = "None", edgecolor = "k", figsize = (20,12), linewidth = 2)
comunemax.to_crs(epsg=3857).plot(ax=ax, facecolor = "None", edgecolor = "k", linewidth = 2)
comunemin.to_crs(epsg=3857).plot(ax=ax, facecolor = "None", edgecolor = "k", linewidth = 2)
ctx.add_basemap(ax)
```

3.  dato il nome di un parco, avere la mappa del parco con i comuni che lo ospitano


```python
a = input("Inserisci il nome del parco: ")
b = parchi[parchi.NOME.str.contains(a)]
c = comuni[comuni.intersects(b.geometry.squeeze())]

ax = b.to_crs(epsg=3857).plot(figsize = (20,12), facecolor = "green", alpha = 0.5)
c.to_crs(epsg=3857).plot(ax=ax, facecolor = "None", edgecolor = "k")
ctx.add_basemap(ax)
```
