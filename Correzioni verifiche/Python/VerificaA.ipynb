scrivere un programma che permetta di verificare se un orario è scritto correttamente. Un orario è scritto correttamente se è formato da 5 caratteri di cui il terzo è il carattere : . Ad esempio 13:45 è scritto correttamente 123:4 non è scritto correttamente.

```python
orario = input("Inserire l'orario: ")
if len(orario)==5:
  if orario[2]==":":
    print("Orario corretto")
  else:
    print("Orario sbagliato")
else:
  print("Orario sbagliato")
```

scrivere una funzione chiamata prodotto che data una lista di numeri interi restituisca il prodotto di tutti i suoi elementi. Ad esempio se la lista fosse [2, 5, 3] la funzione restituirebbe 30


```python
def prodotto (l):
  i = 1
  for e in l:
    i = i*e
  return i

lst = [int(elemento) for elemento in input("Inserisci i numeri: ").split()]
prodotto(lst)
```

scrivere un programma che, utilizzando una lambda function e la funzione map, elevi alla terza tutti gli elementi della lista. La lista deve essere caricata dall’utente.


```python
lst = [int(elemento) for elemento in input("Inserisci i numeri: ").split()]
lstf = list(map(lambda x: x**3, lst))
print(lst)
print(lstf)
```

scrivere una funzione che calcoli in modo ricorsivo il prodotto tra due numeri.


```python
def prodotto2(f1,f2):
  if f2 == 1:
    return f1
  else:
    return f1+prodotto2(f1,f2-1)

prodotto2(5,20)
```
